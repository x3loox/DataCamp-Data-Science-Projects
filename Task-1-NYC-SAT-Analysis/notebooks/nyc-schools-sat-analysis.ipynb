Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()

# Start coding here...
# Add as many cells as you like...
```

**_Which NYC schools have the best math results?_**

```python
best_math_schools = schools[["school_name", "average_math"]].sort_values("average_math", ascending=False)
best_math_schools = best_math_schools[best_math_schools["average_math"] >= 640]
print(best_math_schools)
```

**_What are the top 10 performing schools based on the combined SAT scores?_**

```python
top_10_schools = schools[["school_name"]]
top_10_schools["total_SAT"] = schools[["average_math", "average_reading", "average_writing"]].sum(axis=1)
top_10_schools = top_10_schools.sort_values("total_SAT", ascending=False)
top_10_schools = top_10_schools.head(10)
print(top_10_schools)
```

**_Which single borough has the largest standard deviation in the combined SAT score?_**

```python
num_schools_df = schools.groupby(["borough"])[["school_name"]].count().reset_index()
num_schools_df = num_schools_df.rename(columns={"school_name": "num_schools"})

schools["total_SAT"] = schools[["average_math", "average_reading", "average_writing"]].sum(axis=1)
sat_stats = schools.groupby("borough")["total_SAT"].agg(["mean", "std"]).round(2)
sat_stats = sat_stats.rename(columns={"mean": "average_SAT", "std": "std_SAT"})

largest_std_dev = pd.merge(num_schools_df, sat_stats, on="borough")

largest_std_dev = largest_std_dev.sort_values("std_SAT", ascending=False).head(1)

print(largest_std_dev)
```
